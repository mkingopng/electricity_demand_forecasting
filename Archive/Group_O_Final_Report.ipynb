### Data Preparation

#### Data Ingestion
The raw data collected from different sources as mentioned in the above section needed to be ingested and combined into a single data frame for modelling to be performed on this data. To facilitate this each data set was loaded into a different data frame using ingestion jobs specifically developed for them. 
This included:
> -Scanning source folders for specific file formats (e.g. .csv)

> -Loading files in chunks and combining into a single data frame per data set.
Once each data set is established, the data was then moved into the pre-processing stage.


### Data Cleaning
The key activities undertaken during data cleansing involved detection and handling of missing values, detection and handling of outliers.

#### Missing Values

The detection and handling of missing or null values within each dataset is a crucial step which needs to be conducted in any project. There are a few levels of checks which are performed for the datasets:

##### NaN Data Points
Data points in the datasets which are N/A or null were removed from the dataset using inbuilt pandas functions. 

##### Missing Observations
Observations for days which might be missing from the dataset or data which is not at the correct level of granularity (or contain the correct number of observations per day) and would create some noise in the modelling.
The data being modelled is aggregated from 5-minute intervals into 30-minute intervals. As part of this check we needed to ensure that each day’s observation contained the correct number of observations, and where the correct number of observations were not available for the day, this data was omitted from further analysis and removed from the dataset.
The check of missing daily observations was conducted on aggregated data (i.e. 30-minute intervals), and data points for dates which did not have the correct number of days per date were removed from the overall data set.



1. Missing data: 01-02-2013,17-07-2016,18-07-2016  - data is missing for these dates where there are no daytime. Entire dataframe checked by splitting datetime, grouping on date and counting the timestamps.
2. Outliers:
Outliers found in TEMPERATURE (569), rrp(151) and TOTALDEMAND (151). Outliers identified using:
1. Scatter and Boxplots
2. Z-Score to determine upper and lower limits
3. Isolation Forest

The approach to threshold for TEMPERATURE and rrp were to use capping, effectively changing the value of the datapoint to either the upper or lower limit that was found.TOTALDEMAND was not capped but left as it is. This is our key parameter and we did not want to cap these values.
