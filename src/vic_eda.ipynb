## merge

```python
merged_df = pd.merge(vic_total_demand, vic_forecast_demand, on='DATETIME', how='outer')
merged_df.head()
```

```python
vic_df = pd.merge(merged_df, vic_temperature, on='DATETIME', how='outer')
vic_df.head()
```

## Plots

```python
plt.figure(figsize=(14, 7))
plt.plot(vic_df['DATETIME'], vic_df['FORECASTDEMAND'], label='Forecast Demand')
plt.plot(vic_df['DATETIME'], vic_df['TOTALDEMAND'], label='Actual Demand', alpha=0.7)
plt.xlabel('Datetime')
plt.ylabel('Price')
plt.title('Forecast vs Actual Price Over Time')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(vic_df['DATETIME'], vic_df['TEMPERATURE'], color='orange', label='TEMPERATURE')
plt.xlabel('Datetime')
plt.ylabel('Temperature')
plt.title('Temperature Over Time')
plt.legend()
plt.show()
```

## Scatter Plot

```python
plt.figure(figsize=(10, 6))
plt.scatter(vic_df['TEMPERATURE'], vic_df['TOTALDEMAND'], label='Actual Demand', alpha=0.5)
plt.scatter(vic_df['TEMPERATURE'], vic_df['FORECASTDEMAND'], label='Forecast Demand', alpha=0.5)
plt.xlabel('Temperature')
plt.ylabel('Demand')
plt.title('Temperature vs Demand')
plt.legend()
plt.show()
```

## Box Plot

```python
plt.figure(figsize=(12, 6))
vic_df[['FORECASTDEMAND', 'TOTALDEMAND', 'TEMPERATURE']].plot(kind='box')
plt.title('Distribution of Prices and Temperature')
plt.show()
```

## Histrogram

```python
plt.figure(figsize=(14, 7))
vic_df['TOTALDEMAND'].hist(alpha=0.7, label='Actual Price', bins=20)
vic_df['FORECASTDEMAND'].hist(alpha=0.5, label='Forecast Price', bins=20)
plt.xlabel('Demand')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Demand')
plt.legend()
plt.show()
```

## Histrogram

```python
plt.figure(figsize=(10, 5))
vic_df['TEMPERATURE'].hist(color='orange', bins=20)
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Temperature')
plt.show()
```

## Correlation

```python
plt.figure(figsize=(8, 6))
sns.heatmap(vic_df[['FORECASTDEMAND', 'TOTALDEMAND', 'TEMPERATURE']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
